**Función carga de datos**

```python
import pandas as pd
```

```python
#funcion que se tiene que correr cada vez que vayamos a cargar un nuevo dataset
nombre = input("Escribe el archivo completo que quiere convertir:")

def cargar_dataset (archivo):
    if ".csv" in archivo:
        df = pd.read_csv(archivo)
    elif ".xlsx" in archivo:
        df = pd.read_excel (archivo)
    elif ".json" in archivo:
        df = pd.read_json(archivo)
    elif ".html" in archivo:
        df = pd.read_html(archivo)
        
    #aqui para que nos imprima directamente el .csv que el profe nos pide 
    print(df.to_csv(input("Escribe el nombre del nuevo archivo con .csv al final: ")))
    return (df)
```

**IMPRESIÓN / CSV**

```python
#Francia.csv
#Al correr esta impresion de la función, es cuando me pide un nuevo nombre para crear el csv
cargar_dataset(nombre)
```

**Tuve que volver a hacer el código porque el dataframe de html me lo leía como lista. Entonces seleccioné una posición del dataframe, ya que no me lo quería convertir a .csv**

```python
#funcion que se tiene que correr cada vez que vayamos a cargar un nuevo dataset
nombre = input("Escribe el archivo completo que quiere convertir:")

def cargar_dataset (archivo):
    if ".html" in archivo:
        df_list = pd.read_html(archivo)  # Esto devuelve una lista de dataframes
        df = df_list[0]  # Seleccione el primer dataframe de la lista

    #aqui para que nos imprima directamente el .csv que el profe nos pide 
    print(df.to_csv(input("Escribe el nombre del nuevo archivo con .csv al final: ")))
    return (df)   
```
